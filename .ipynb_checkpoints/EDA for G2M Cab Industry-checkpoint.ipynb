# Exploring the Cab_Data.csv file

```python
Cab_Data_df = pd.read_csv('Cab_Data.csv')
Cab_Data_df.head()
```

```python
print(len(Cab_Data_df['Company']))
print(len(Cab_Data_df['City']))
Cab_Data_df.describe()
```

__As we can see from the describe table anjd the lengths of company and city columns, there are no missing values in Cab_Data.csv__

```python
Cab_Data_df.columns = ['txnID','travel_date','company','city','travel_distance','price','cost']
Cab_Data_df.head()
```

```python
Cab_Data_df['profit'] = Cab_Data_df['price'] - Cab_Data_df['cost']
Cab_Data_df.head()
```

```python
Cab_Data_df.company.value_counts()
```

```python
Cab_Data_df.city.value_counts()
```

```python
print(len(Cab_Data_df[Cab_Data_df['profit'] > 0]))
```

```python
no_profit_df = Cab_Data_df[Cab_Data_df['profit'] <= 0]
no_profit_df.city.value_counts()
```

```python
def cab_separator(df, city):
    city_df = df[df['city'] == city]
    pink = city_df.company.value_counts()['Pink Cab']
    yellow = city_df.company.value_counts()['Yellow Cab']
    return pink,yellow
```

```python
city_list = list(Cab_Data_df['city'].unique())
pink_list = []
yellow_list = []
for city in city_list:
    pink,yellow = cab_separator(Cab_Data_df, city)
    pink_list.append(pink)
    yellow_list.append(yellow)
pink_list
```

```python
x = np.arange(19)
width = 0.40
plt.figure(figsize=(30, 20), dpi=80)
plt.bar(x, yellow_list, width, color='gold')
plt.bar(x, pink_list, width, bottom=yellow_list, color='pink')

plt.xticks(x, city_list, rotation=45, fontsize=15)
plt.yticks(fontsize=20)
plt.xlabel("Cities", fontsize=15)
plt.ylabel('Customers', rotation=360, fontsize=20)
plt.title('Customer Distribution in Different Areas of Operation', fontsize=30)
plt.legend(['Pink','Yellow'], prop={'size':20}, fontsize=15)
```

```python
pink = Cab_Data_df[Cab_Data_df['company'] == 'Pink Cab']
yellow = Cab_Data_df[Cab_Data_df['company'] == 'Yellow Cab']
pink_less = pink[pink['travel_distance'] < 2]
yellow_less = yellow[yellow['travel_distance'] < 2]
print('Pink Less Travelled Avg. Profit: ', pink_less.profit.mean())
print('Yellow Less Travelled Avg. Profit: ', yellow_less.profit.mean(),'\n')
print('Pink Avg. Profit: ', pink.profit.mean())
print('Yellow Avg. Profit: ', yellow.profit.mean(),'\n')
print('Pink Avg. Price: ', pink.price.mean())
print('Yellow Avg. Price: ', yellow.price.mean(),'\n')
print('Pink Avg. Cost: ', pink.cost.mean())
print('Yellow Avg. Cost: ', yellow.cost.mean())
```

__As we can see, the average profit made by the Yellow Company on rides shorter than 2 kms is significantly higher than the average profits made by the Pink Company.__

__In fact, we see this trend continue into the rides of all distances, with the average profit of the Yellow company being almost 2.56 times greater than the average profit earned by the Pink company.__

__By inspecting further the price and cost of both the companies, we observe that while Yellow company has a higher average cost than Pink Company, the difference is less than 50 dollars. The difference in average price, however, is much more significant with the Yellow company's price exceeding that of the Pink company's by almost 140 dollars.__

# Exploring the City.csv data

__Seeing this data, we can see that we are given the population and number of cab users in each city where the data was collected.__

```python
city = pd.read_csv('City.csv')
city.columns = ['city','population','users']
city['population'] = city['population'].str.replace(',', '')
city['users'] = city['users'].str.replace(',', '')
city['population'] = city['population'].apply(lambda x: float(x.strip()))
city['users'] = city['users'].apply(lambda x: float(x.strip()))
city['user_ratio'] = city['users'] / city['population']
city.head()
print(len(df.city.value_counts().index))
print(len(city.city.value_counts().index))
```

__Using this data, we can calculate the user ratio, giving us an idea of how big the market is in each city. This helps us understand the potential in each  city, allowing us to analyze which company has the better infrastructure to expand.__

__We can now use this information as part of our eda. For each row in the relavant data (df), we have a city. We plug in the population and user_ratio values calculated for each city.__

```python
city = city.set_index('city')

pop_list = []
ratio_list = []
pop_value = {}
for index, row in city.iterrows():
    pop_value[index] = [row['population'],row['user_ratio']]

city = city.reset_index()
city_list = list(df['city'])

for city in city_list:
    pop_list.append(pop_value[city][0])
    ratio_list.append(pop_value[city][1])
print(pop_list)


population = pd.Series(pop_list)
user_ratio = pd.Series(ratio_list)
    
df.head()
```
